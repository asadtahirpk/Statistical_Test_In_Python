### Independent T-Test

```python
# Null Hypothesis is Boys and Girls Average marks is same
# Alpha=0.05
import numpy as np
from scipy import stats

# Example data
boys = [78, 85, 82, 90, 76, 88, 84]
girls = [72, 70, 65, 80, 68, 74, 77]

# Independent t-test
t_stat, p_val = stats.ttest_ind(boys, girls)

print("T-statistic:", t_stat)
print("P-value:", p_val)
# Null Hyppothesis is reject after test because P value is less then 0.05 
```

### Paired T-Test

```python
# Null Hypothesis Coaching ka koi effect nahi hai, yani before aur after scores ka mean same hai.
# Alpha=0.05
from scipy import stats

# Example data
before = [65, 70, 68, 72, 66, 71, 69]
after  = [70, 75, 72, 78, 70, 76, 74]

# Paired t-test
t_stat, p_val = stats.ttest_rel(before, after)

print("T-statistic:", t_stat)
print("P-value:", p_val)
# Coaching ka real effect hai → before aur after scores me significant difference hai
```

### ANOVA TEST 

```python
# One Way:
# Null Hypothesis (H₀): Sab groups ka mean same hai → koi real difference nahi
from scipy import stats

# Example data: 3 teaching methods
method_A = [85, 88, 90, 86, 87]
method_B = [78, 75, 80, 74, 77]
method_C = [92, 95, 91, 94, 90]

# One-way ANOVA
f_stat, p_val = stats.f_oneway(method_A, method_B, method_C)

print("F-statistic:", f_stat)
print("P-value:", p_val)
# P-value bohot chhoti hai (< 0.05), isliye null hypothesis (H₀) reject karte hain.
# Matlab: At least ek teaching method ka mean score doosre method se significantly different hai.
```

```python
# Two Way 
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Example data
data = {
    'Score': [85, 88, 90, 86, 87, 78, 75, 80, 74, 77,
              92, 95, 91, 94, 90, 88, 85, 87, 84, 86],
    'Method': ['A','A','A','A','A','B','B','B','B','B',
               'C','C','C','C','C','A','B','C','A','B'],
    'Gender': ['M','F','M','F','M','M','F','M','F','M',
               'F','M','F','M','F','F','F','M','M','F']
}

df = pd.DataFrame(data)

# Two-Way ANOVA
model = ols('Score ~ C(Method) + C(Gender) + C(Method):C(Gender)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

df=pd.DataFrame(anova_table)
df
```

Teaching Method ka students ke scores par significant effect hai (p < 0.05).
→ Different methods se students ke scores mein farq aata hai.

Gender ka scores par koi significant effect nahi hai (p > 0.05).
→ Male aur Female students ke average scores similar hain.

Interaction (Method × Gender) bhi significant nahi hai (p > 0.05).
→ Matlab teaching method ka effect gender ke hisaab se change nahi hota.

### Pearson Corelation

```python
import numpy as np
from scipy import stats
import seaborn as sns 
import matplotlib.pyplot as pyt 
# Data
hours = np.array([2, 4, 5, 7, 9])
scores = np.array([50, 55, 65, 70, 88])

# Pearson correlation
r, p_value = stats.pearsonr(hours, scores)

print("Correlation Coefficient (r):", r)
print("P-value:", p_value)
sns.set_style("darkgrid")
sns.scatterplot(x=hours,y=scores,color="blue")
slope, intercept, r_value, p_value, std_err = stats.linregress(hours,scores)
regression_line = [slope * xi + intercept for xi in hours]

pyt.plot(hours, regression_line, color="red")
pyt.title(f"Strong positive corelation r={r}",color="red")
pyt.xlabel("Study Hours",color="red")
pyt.ylabel("Scores Gain",color="red")



```

### Simple Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Sample Data (Hours Studied vs Exam Score)
x = np.array([1, 2, 3, 4, 5, 6, 7, 8])   # Hours studied
y = np.array([35, 40, 50, 55, 65, 70, 75, 85])  # Exam score

# Regression Analysis
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

print("Slope (β₁):", slope)
print("Intercept (β₀):", intercept)
print("R-squared:", r_value**2)

# Predicted values
regression_line = [slope * xi + intercept for xi in x]

# Plot
plt.scatter(x, y, color="blue", label="Actual Data")   # real data
plt.plot(x, regression_line, color="red", label="Regression Line")  # regression line
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.legend()
plt.show()
```

###  Chi Square Goodness of Fit Test

Scenario:
Ek dice ko 60 dafa fainka.
Agar dice fair ho to har number (1–6) approx 10 times aana chahiye.
Lekin actual result ye mila:

1 → 8

2 → 9

3 → 12

4 → 11

5 → 10

6 → 10

```python
from scipy.stats import chisquare

observed = [8, 9, 12, 11, 10, 10]
expected = [10, 10, 10, 10, 10, 10]

chi2, p = chisquare(f_obs=observed, f_exp=expected)
print("Chi-Square:", chi2)
print("P-value:", p)
```

It mean observed RESULT and expected result are same because p value is greater then 0.05 ans null hypohesis is not  rejected 

### Chi Square Test of Independence

Scenario:
Hum dekhna chahte hain ke Gender aur Smoking ke darmiyan koi relation hai ya nahi.

Data Table (Observed):
|           | Smoking Yes | Smoking No | Total |
| --------- | ----------- | ---------- | ----- |
| Male      | 40          | 20         | 60    |
| Female    | 10          | 30         | 40    |
| **Total** | 50          | 50         | 100   |


```python
import numpy as np
from scipy.stats import chi2_contingency

observed = np.array([[40, 20],
                     [10, 30]])

chi2, p, dof, expected = chi2_contingency(observed)
print("Chi-Square:", chi2)
print("P-value:", p)
print("Expected Table:\n", expected)
```

p is less then 0.05 it means gender and smoking categories are dependent variables 

### Mann–Whitney U Test (Independent Samples)

Kya male aur female students ke exam scores mein fark hai?

```python
import numpy as np
from scipy.stats import mannwhitneyu

# Scores of two independent groups
male_scores = [85, 90, 78, 92, 88]
female_scores = [70, 75, 80, 68, 72]

# Mann-Whitney U Test
stat, p = mannwhitneyu(male_scores, female_scores)

print("Mann-Whitney U Statistic:", stat)
print("P-value:", p)
```

P<0.05 it means dono groups ky score mai significant diference h 

### Wilcoxon Signed-Rank Test (Paired Samples)

Kya diet plan ke baad students ka weight change hua?

```python
from scipy.stats import wilcoxon

# Before and after weights of same students
before = [70, 80, 90, 85, 95]
after  = [68, 78, 88, 83, 92]

# Wilcoxon Signed-Rank Test
stat, p = wilcoxon(before, after)

print("Wilcoxon Statistic:", stat)
print("P-value:", p)
```

Null hypothesis is not rejected beacuse p is greater then 0.05

### Kruskal–Wallis Test (3+ Independent Groups)

Kya 3 teaching methods ke students ke scores mein fark hai?

```python
from scipy.stats import kruskal

# Scores from three independent groups
method_A = [85, 80, 82, 90, 88]
method_B = [70, 72, 68, 75, 74]
method_C = [60, 65, 63, 58, 62]

# Kruskal-Wallis Test
stat, p = kruskal(method_A, method_B, method_C)

print("Kruskal-Wallis Statistic:", stat)
print("P-value:", p)
```

p is less then 0.05 it means Null hypothesis is  rejected and  3 diferent teaching method mai farq h 

### Friedman Test (3+ Related Groups)

Kya same students ka performance 3 subjects mein different hai?

```python
from scipy.stats import friedmanchisquare

# Scores of same students in 3 subjects
math    = [85, 78, 92, 88, 80]
science = [82, 76, 90, 85, 79]
english = [88, 80, 95, 90, 83]

# Friedman Test
stat, p = friedmanchisquare(math, science, english)

print("Friedman Statistic:", stat)
print("P-value:", p)
```

It means diferent is significant

### Spearman’s Rank Correlation

```python
import numpy as np
from scipy.stats import spearmanr

# Student attendance (in %)
attendance = [60, 70, 80, 90, 95, 85, 75, 65]

# Exam scores (out of 100)
scores = [55, 65, 78, 88, 92, 85, 70, 60]

# Spearman's Rank Correlation
corr, p = spearmanr(attendance, scores)

print("Spearman Correlation Coefficient:", corr)
print("P-value:", p)
```

coefficient is 1 its mean prefect positive corelation
